```python
# step 1: build sigmoid function:
def sigmoid(z):
    g = 1/(1+np.exp(-z))
    return g
print(sigmoid(np.array([-1,0,1,2])))
```

```python
#step 2: build loss function
def CostFunction(X_data,y_data,w,b): # trong do 
    # w la ma tran 0 co do dai la so cot cua x
    cost = 0
    for i in range(len(y_data)):
        # co nghia la w1*x1 + w2*x2 + b
        fwb = np.dot(w,X_data[i,:])+b # ma tran 100,1
        fwb_i = sigmoid(fwb)
        loss = (-y_data[i]*np.log(fwb_i)) - (1-y_data[i])*np.log(1-fwb_i)
        cost += loss
    cost = 1/(len(y_data)) * cost
    return cost
m,n = X_data.shape;
w = np.zeros(n)
b = 0
print(CostFunction(X_data,y_data,w,b))
```

```python
# step 3: gradient descent for logistic regression
def GradientDescent(X_data,y_data,w,b):
    
    dj_dw = np.zeros(w.shape) # 2,1
    dj_db = 0
    for i in range(m):
        zwb = 0
        for j in range(n):
            zwb += w[j]*X_data[i][j] # 1 
        # after for: 100 gia tri duoc chay!
        zwb += b # 1
        fwb = sigmoid(zwb)
        dj_db += fwb - y_data[i]
        for j in range(n):
            dj_dw[j] += (fwb-y_data[i])*X_data[i][j]
        
    dj_dw = (1/m) * dj_dw
    dj_db = (1/m) * dj_db
    return dj_dw, dj_db

print(GradientDescent(X_data,y_data,np.array([0.2,-0.5]),-24))
    
```

```python
J_history = []
w_history = []
alpha = 0.001
def LearningAlgor(X_data,y_data,w,b):
    for i in range(10000):
        dj_dw, dj_db = GradientDescent(X_data,y_data,w,b)
        w = w - alpha*dj_dw
        b = b - alpha*dj_db
        J_history.append(CostFunction(X_data,y_data,w,b))
        w_history.append(w)
    return w,b
w_in, b_in = LearningAlgor(X_data,y_data,w,b)
print(w_in)
print(b_in)
print(J_history)
```
